## stationarity, moving averages & autocorrelations
- Check for stationarity
.rolling
.shift

```python
#check for stationarity
ad_fuller_result1 = adfuller(y_train)
print("p-value for original: ", str(ad_fuller_result1[1]))

dengue_cases = y_train.diff().dropna()
ad_fuller_result2 = adfuller(dengue_cases)
print("p-value for diff: ", str(ad_fuller_result2[1]))

plot_acf(y_train, lags = 52, title = 'Autocorrelation of dengue cases');
plot_pacf(y_train, lags = 52, title = 'Partial autocorrelation of dengue cases');
```

```python
#check for stationarity
#instantiate
adftest = pm.arima.ADFTest(alpha = 0.05)

#original
adftest_result1 = adftest.should_diff(y_train)
print("p-value for original: ", str(adftest_result1[0]))
#diff
dengue_cases = y_train.diff().dropna()
adftest_result2 = adftest.should_diff(dengue_cases)
print("p-value for diff: ", str(adftest_result2[0]))

plot_acf(y_train, lags = 52, title = 'Autocorrelation of dengue cases');
plot_pacf(y_train, lags = 52, title = 'Partial autocorrelation of dengue cases');
```

```python
from statsmodels.tsa.stattools import acf 
from statsmodels.tsa.stattools import pacf 

acf_corel_list = []
for n in acf(dengue_cases, nlags = 52):
    if n > 0.05:
        acf_corel_list.append(round(n,2))

pacf_corel_list = []
for n in pacf(dengue_cases, nlags = 52):
    if n > 0.05:
        pacf_corel_list.append(round(n,2))

    
acf_corel_list, pacf_corel_list
```

Since p-value is less the 0.05 even for the orignal time series, the null hypthesis that the data is non-stationary is rejected and the alternative hypothesis that the data is stationary can be accepted. With stationarity confirmed at the 95% signficance level, we can proceed with $d$ = 0.
<br>

The autocorrelation plot checks correlation between itself and all previous data points (lags). 1st correlation coefficient '1' is ignored since it represents correlation with itself which is perfect (1). However, the lags shows a continuous curve which indicates that the data is highly correlated and it is difficult to show any seasonality.

This is where the Partial autocrrelation plots comes in since, removing the effect of the correlations due to the terms at shorter lags. However, there isnt really much seasonality except for a lag of 1 (week), even though there seems to be some inverser correlations at lag 3,4, with the rest not considered statistically significant or being borderline cases.

~The 3rd, 4th, 5th, 9th, 12th and  14th points outside the shaded area _(95% confidence interval)_ are considered statistically significant, while those that are within are not. These means that the points are correlation coefficients of different lags in the dengue data, with correlation coefficients of 0.30, 0.15, 0.13, 0.07, 0.06, 0.05 respectively. Thus, it is quite apparent that the strongest lag time correlations is about 3 weeks, followed by 5 weeks & 4 weeks, 12 weeks and~

## Scoring metrics

 - Mean Average Percentage Error (MAPE) is typically used as a metric in time series due to the value of its relatively scale against the observations/predictions. This makes it convenient for comparison althought the large variations in the actual cases and predictions may not translate well to the actual number of cases.
 - Mean Average Error (MAE) can be used as a metric to measure the absolute distance (differences in cases) of the observations against the predictions. The absolute value would account for negative errors as well.
 - Root Mean Squared Error (RMSE) is also used in conjuction to help track the robustness of the model. Having the errors squared also accounts for negative errors and serves as additional penalty against higher errors.
     - Large differences between MAE and RMSE would highlight that the model would likely be plagued by outliers that may be affecting its fit.

### Functions
* For simplifying plotting 
* For metrics evaluations

```python
def score_metric2(train, test, pred_train, pred_test):
    mape_train = mean_absolute_percentage_error(train, pred_train)
    mape_test = mean_absolute_percentage_error(test, pred_test)
    mae_train = mean_absolute_error(train, pred_train)
    mae_test = mean_absolute_error(test, pred_test)
    rmse_train = mean_squared_error(train, pred_train, squared = False)
    rmse_test = mean_squared_error(test, pred_test, squared = False)
    print(f"Mean Absolute Percentage Error (MAPE), training: {mape_train:.4f}")
    print(f"Mean Absolute Percentage Error (MAPE), testing: {mape_test:.4f}")
    print(f"Mean Absolute Error (MAE), training: {mae_train:.4f}")
    print(f"Mean Absolute Error (MAE), testing: {mae_test:.4f}")
    print(f"Root Mean Squared Error (RMSE), training: {rmse_train:.4f}")
    print(f"Root Mean Squared Error (RMSE), testing: {rmse_test:.4f}")
```

```python
#for generating the score_metric
def score_metric(test, pred):
    mape = mean_absolute_percentage_error(test, pred)
    mae = mean_absolute_error(test, pred)
    rmse = mean_squared_error(test, pred, squared = False)
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    return mape, mae, rmse
```

```python
# Define function to help make predictions and plot the results 
def arima_forecast(y_train, y_test, arima_model, y_period, X_test = None):
    # default None for X_test to share function between (S)ARIMA(X),    
    # Generate len(y_test) number of predictions
    if X_test is None:
        y_pred = arima_model.predict(n_periods = y_period) 
    else:
        y_pred = arima_model.predict(n_periods = y_period, X = X_test[:y_period]) 
    
    # Plot data.
    plt.figure(figsize=(30,15))

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_train, color='tab:blue', label='y_train')

    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_test, color='orange', label='y_test', alpha = 0.8, ls = '--')

    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_pred, color='green', label='y_pred', alpha = 0.8, ls = '--')
    
    #MAPE (mean abs percentage error) of the predictions for quick glance 
#    mae = mean_absolute_error(y_test, y_pred) # like all sklearn metrics we've used, pass y_actual, y_pred
    mape, mae, rmse = score_metric(y_test[:y_period], y_pred)
    
    plt.title(label = f'Sales Predictions\n MAPE: {mape:.2f}\n MAE: {mae:.2f}', fontsize=16) 
    plt.legend(fontsize = 20, loc = 'upper left');
```

```python
# Define function to help make predictions and plot the results 
def model_forecast(X_train, X_test, y_train, y_test, gridsearch_model):
    # default None for X_test to share function between (S)ARIMA(X),
    # Generate len(y_test) number of predictions
    train_pred = gridsearch_model.predict(X_train)
    test_pred = gridsearch_model.predict(X_test)
    
    # Plot data.
    plt.figure(figsize=(30,15))

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_train, color='tab:blue', label='y_train')

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(train_pred, color='tab:cyan', label='y_train')
    
    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_test, color='orange', label='y_test', alpha = 0.8, ls = '--')

    # Plot predicted test values. as shared previously, there are no index in y_pred 
    sns.lineplot(test_pred, color='tab:green', label='y_pred', alpha = 0.8, ls = '-.')
    
    #MAPE (mean abs percentage error) of the predictions for quick glance 
#    mae = mean_absolute_error(y_test, y_pred) # like all sklearn metrics we've used, pass y_actual, y_pred
    print(f"Score metrics of train data:\n{'-'*40}")
    mape, mae, rmse = score_metric(y_train, train_pred)
    print(f"\n{'_'*50}\n Score metrics of test data:\n{'-'*40}")
    mape, mae, rmse = score_metric(y_test, test_pred)
    
    plt.title(label = f'Sales Predictions\n MAPE: {mape:.2f}\n MAE: {mae:.2f}', fontsize=16) 
    plt.legend(fontsize = 20, loc = 'upper left');
```
