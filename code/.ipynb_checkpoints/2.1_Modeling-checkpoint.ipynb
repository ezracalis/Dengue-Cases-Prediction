## stationarity, moving averages & autocorrelations
- Check for stationarity
.rolling
.shift

```python
#check for stationarity
ad_fuller_result1 = adfuller(dengue['dengue_cases'])
print("p-value for original: ", str(ad_fuller_result1[1]))

dengue_cases = dengue['dengue_cases'].diff().dropna()
ad_fuller_result2 = adfuller(dengue_cases)
print("p-value for diff: ", str(ad_fuller_result2[1]))

plot_acf(dengue_cases, lags = 52, title = 'Autocorrelation of dengue cases');
plot_pacf(dengue_cases, lags = 52, title = 'Partial autocorrelation of dengue cases');
```

```python
#check for stationarity
#instantiate
adftest = pm.arima.ADFTest(alpha = 0.05)

#original
adftest_result1 = adftest.should_diff(dengue['dengue_cases'])
print("p-value for original: ", str(adftest_result1[0]))
#diff
dengue_cases = dengue['dengue_cases'].diff().dropna()
adftest_result2 = adftest.should_diff(dengue_cases)
print("p-value for original: ", str(adftest_result2[0]))

plot_acf(dengue_cases, lags = 52, title = 'Autocorrelation of dengue cases');
plot_pacf(dengue_cases, lags = 52, title = 'Partial autocorrelation of dengue cases');
```

```python
from statsmodels.tsa.stattools import acf 
from statsmodels.tsa.stattools import pacf 

acf_corel_list = []
for n in acf(dengue_cases, nlags = 52):
    if n > 0.05:
        acf_corel_list.append(round(n,2))

pacf_corel_list = []
for n in pacf(dengue_cases, nlags = 52):
    if n > 0.05:
        pacf_corel_list.append(round(n,2))

    
acf_corel_list, pacf_corel_list
```

Since p-value is less the 0.05 even for the orignal time series, the null hypthesis that the data is non-stationary is rejected and the alternative hypothesis that the data is stationary can be accepted. With stationarity confirmed at the 95% signficance level, we can proceed with $d$ = 0.
<br>

The autocorrelation plot checks correlation between itself and all previous data points (lags). 1st correlation coefficient '1' is ignored since it represents correlation with itself which is perfect (1).

The 3rd, 4th, 5th, 9th, 12th and  14th points outside the shaded area _(95% confidence interval)_ are considered statistically significant, while those that are within are not. These means that the points are correlation coefficients of different lags in the dengue data, with correlation coefficients of 0.31, 0.13, 0.15, 0.06, 0.09, 0.05 respectively. ## Scoring metrics

 - Mean Average Error (MAE) can be used as a metric to measure the absolute distance (differences in cases) of the observations against the predictions. The absolute value would account for negative errors as well.
     -  While Mean Average Percentage Error (MAPE) is typically used as a metric in time series due to the value of its relatively scale against the observations/predictions, the large variations in the predictions (dengue cases may even be 0) may cause the error to increase when the actual value of the predictions are low, likely skewing the results. This makes its unsuitable.
 - Root Mean Squared Error (RMSE) is also used in conjuction to help track the robustness of the model. Having the errors squared also accounts for negative errors and serves as additional penalty against higher errors.
     - Large differences between MAE and RMSE would highlight that the model would likely be plagued by outliers that may be affecti its fit.

### Functions
* For simplifying plotting 
* For metrics evaluations

```python
def score_metric(train, test, pred_train, pred_test):
    mape_train = mean_absolute_percentage_error(train, pred_train)
    mape_test = mean_absolute_percentage_error(test, pred_test)
    mae_train = mean_absolute_error(train, pred_train)
    mae_test = mean_absolute_error(test, pred_test)
    rmse_train = mean_squared_error(train, pred_train, squared = False)
    rmse_test = mean_squared_error(test, pred_test, squared = False)
    print(f"Mean Absolute Percentage Error (MAPE), training: {mape_train:.4f}")
    print(f"Mean Absolute Percentage Error (MAPE), testing: {mape_test:.4f}")
    print(f"Mean Absolute Error (MAE), training: {mae_train:.4f}")
    print(f"Mean Absolute Error (MAE), testing: {mae_test:.4f}")
    print(f"Root Mean Squared Error (RMSE), training: {rmse_train:.4f}")
    print(f"Root Mean Squared Error (RMSE), testing: {rmse_test:.4f}")
```

```python
#for generating the score_metric
def score_metric(test, pred):
    mape = mean_absolute_percentage_error(test, pred)
    mae = mean_absolute_error(test, pred)
    rmse = mean_squared_error(test, pred, squared = False)
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    return mape, mae, rmse
```

```python
# Define function to help make predictions and plot the results 
def arima_forecast(y_train, y_test, arima_model, X_test=None):
    # default None for X_test to share function between (S)ARIMA(X),    
    # Generate len(y_test) number of predictions
    y_pred = arima_model.predict(n_periods = len(y_test), X=X_test) # we want as many future data point predictions as there are in y_test

    # Plot data.
    plt.figure(figsize=(30,15))

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_train, color='tab:blue', label='y_train')

    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_test, color='orange', label='y_test', alpha = 0.8, ls = '--')

    # Plot predicted test values. as shared previously, there are no index in y_pred 
    sns.lineplot(y_pred, color='tab:green', label='y_pred', alpha = 0.8, ls = '-.')
    
    #MAPE (mean abs percentage error) of the predictions for quick glance 
#    mae = mean_absolute_error(y_test, y_pred) # like all sklearn metrics we've used, pass y_actual, y_pred
    mape, mae, rmse = score_metric(y_test, y_pred)
    
    plt.title(label = f'Sales Predictions\n MAE: {mae:.2f}', fontsize=16) 
    plt.legend(fontsize = 20, loc = 'upper left');
```

```python
# Define function to help make predictions and plot the results 
def model_forecast(X_train, X_test, y_train, y_test, gridsearch_model):
    # default None for X_test to share function between (S)ARIMA(X),
    # Generate len(y_test) number of predictions
    train_pred = gridsearch_model.predict(X_train)
    test_pred = gridsearch_model.predict(X_test)
    
    # Plot data.
    plt.figure(figsize=(30,15))

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_train, color='tab:blue', label='y_train')

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(train_pred, color='tab:cyan', label='y_train')
    
    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_test, color='orange', label='y_test', alpha = 0.8, ls = '--')

    # Plot predicted test values. as shared previously, there are no index in y_pred 
    sns.lineplot(test_pred, color='tab:green', label='y_pred', alpha = 0.8, ls = '-.')
    
    #MAPE (mean abs percentage error) of the predictions for quick glance 
#    mae = mean_absolute_error(y_test, y_pred) # like all sklearn metrics we've used, pass y_actual, y_pred
    print(f"Score metrics of train data:\n{'-'*40}")
    mape, mae, rmse = score_metric(y_train, train_pred)
    print(f"\n{'_'*50}\n Score metrics of test data:\n{'-'*40}")
    mape, mae, rmse = score_metric(y_test, test_pred)
    
    plt.title(label = f'Sales Predictions\n MAE: {mae:.2f}', fontsize=16) 
    plt.legend(fontsize = 20, loc = 'upper left');
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, LSTM


# Define X (features) and y (target)
X = dengue.drop(columns=['dengue_cases','search_dengue_symptoms',
            'daily_rainfall_total_mm',
            'mean_temp',
            'max_temp',
            'min_temp',
            'mean_wind_speed_kmh'])
y = dengue['dengue_cases']

# Normalize the data
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))

# Split data into training and testing sets based on date
X_train = X_scaled[dengue.index <= '2021']
X_test = X_scaled[(dengue.index >= '2022-01') & (dengue.index <= '2022-12')]
y_train = y_scaled[dengue.index <= '2021']
y_test = y_scaled[(dengue.index >= '2022-01') & (dengue.index <= '2022-12')]

# Reshape the data for LSTM
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
model = Sequential()

model.add(LSTM(128, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(LSTM(64, return_sequences=True))
model.add(LSTM(16))

model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation = 'linear'))

model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), shuffle=False)
```

```python
# Make predictions
train_pred = model.predict(X_train)
y_pred = model.predict(X_test)

# Transform predictions back to the original scale
y_pred_original = scaler_y.inverse_transform(y_pred)
y_test_original = scaler_y.inverse_transform(y_test)
train_pred_original = scaler_y.inverse_transform(train_pred)
y_train_original = scaler_y.inverse_transform(y_train)
```
