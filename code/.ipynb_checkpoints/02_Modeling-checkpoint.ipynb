### Scoring Metrics

 - Mean Average Percentage Error (MAPE) is typically used as a metric in time series due to the value of its relative scale against the observations/predictions. This makes it convenient for comparison although the large variations in the actual cases and predictions may not translate well to the actual number of cases.
 - Mean Average Error (MAE) can be used as a metric to measure the absolute distance (differences in cases) of the observations against the predictions. The absolute value would account for negative errors as well.
 - Root Mean Squared Error (RMSE) is also used in conjuction to help track the robustness of the model. Having the errors squared also accounts for negative errors and serves as additional penalty against higher errors.
     - Large differences between MAE and RMSE would highlight that the model would likely be plagued by outliers that may be affecting its fit.

Thus, our primary metric will be MAPE, with MAE and RMSE being the secondary metrics.

### Functions

We will create a few functions to simplify the evaluation of metrics and model plotting.

````python
def score_metric(train, test, pred_train, pred_test):
    mape_train = mean_absolute_percentage_error(train, pred_train)
    mape_test = mean_absolute_percentage_error(test, pred_test)
    mae_train = mean_absolute_error(train, pred_train)
    mae_test = mean_absolute_error(test, pred_test)
    rmse_train = mean_squared_error(train, pred_train, squared = False)
    rmse_test = mean_squared_error(test, pred_test, squared = False)
    print(f"Mean Absolute Percentage Error (MAPE), training: {mape_train:.4f}")
    print(f"Mean Absolute Percentage Error (MAPE), testing: {mape_test:.4f}")
    print(f"Mean Absolute Error (MAE), training: {mae_train:.4f}")
    print(f"Mean Absolute Error (MAE), testing: {mae_test:.4f}")
    print(f"Root Mean Squared Error (RMSE), training: {rmse_train:.4f}")
    print(f"Root Mean Squared Error (RMSE), testing: {rmse_test:.4f}")
````

````python
# Define function to help make predictions and plot the results 
def arima_forecast(y_train, y_test, arima_model, y_period, X_test = None):
    # default None for X_test to share function between (S)ARIMA(X),    
    # Generate len(y_test) number of predictions
    if X_test is None:
        y_pred = arima_model.predict(n_periods = y_period) 
    else:
        y_pred = arima_model.predict(n_periods = y_period, X = X_test[:y_period]) 
    
    # Plot data.
    plt.figure(figsize=(30,15))

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_train, color='tab:blue', label='y_train')

    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_test, color='orange', label='y_test', alpha = 0.8, ls = '--')

    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    sns.lineplot(y_pred, color='green', label='y_pred', alpha = 0.8, ls = '--')
    
    #MAPE (mean abs percentage error) of the predictions for quick glance 
#    mae = mean_absolute_error(y_test, y_pred) # like all sklearn metrics we've used, pass y_actual, y_pred
    mape_test = mean_absolute_percentage_error(y_test[:y_period], y_pred)
    
    plt.title(label = f'Dengue predictions \n MAPE: {mape_test:.2f}', fontsize=16) 
    plt.legend(fontsize = 20, loc = 'upper left');
````

### ARIMA

#### Modeling

````python
#assign X & y and straitfy the imbalance observed during cleaning
X = dengue[['search_dengue', 
            'search_dengue_symptoms',
            'daily_rainfall_total_mm',
            'mean_temp',
            'max_temp',
            'min_temp',
            'mean_wind_speed_kmh',
            'max_wind_speed_kmh']]
y = dengue['dengue_cases']

# split `X` train and test for same timeframe as `y` 
# Test: Nov 2022 onwards 
X_train = X[:'2020']
X_test = X['2021-01':'2021-03']
y_train = y[:'2020']
y_test = y['2021-01':'2021-03']

#check
print(X_train.iloc[:,0].tail()) # confirm ending at 2020 for train
print(X_test.iloc[:,0].head()) # confirm beginning at 2021 for test
````

````python
# Fit Auto ARIMA
arima_model = pm.AutoARIMA(start_p=0, max_p=10, # find optimum value of p automatically
                           d=None,              # find optimum value of d automatically
                           start_q=0, max_q=10, # find optimum value of q automatically
                           trace=True,          # Print values for each fit in the grid search
                           random_state=42,
                           n_fits=50,
                           test = 'adf'
                          )
arima_model.fit(y_train)
````

````python
pred_train = arima_model.predict_in_sample() 
pred_test = arima_model.predict(len(y_test))
score_metric(y_train, y_test, pred_train, pred_test)
````

````python
arima_forecast(y_train, y_test, arima_model, y_period = len(y_test))
````

*Interpretation: From the above plot, it is quite evident that the ARIMA model is averaging the data points and pushing the prediction higher, leading to a rather high test MAPE of 41%. We also see signs of overfitting as the train scores are much better than the test scores.*

### Dengue Prevention Programs 

Over the past few decades, dengue prevention programs has evolved to incorporate new technologies to prevent dengue infection. The widely known dengue prevention project currently in  large scale deployment is the "Project Wolbachia" program. This is also known as the incompatible insect technique (IIT) whereby male _Aedes_ mosquitoes infected with the Wolbachia bacteria is released to breed with the wild-type female _Aedes aegypti_ mosquito to yield non-viable eggs, thereby reducing the number of _Aedes_ mosquitoes in the community. The project has been shown to lower the population _Aedes_ mosquitoes by up to 98% as well as reducing the number of dengue fever cases by up to 88% [(Reference)](https://www.medrxiv.org/content/10.1101/2021.06.16.21257922v1). Currently the project is being conducted in 13 estates around Singapore which coverage of nearly 300,000 households by the Housing Develoment Board (HDB) flats[(Reference)](https://www.nea.gov.sg/corporate-functions/resources/research/wolbachia-aedes-mosquito-suppression-strategy/wolbachia-aedes-release-schedule). It was estimated by Soh et. al that to reach a steady state of covering the entire Singapore, project Wolbachia would cost the economy around SGD 40,000,000 per year which would be approximately SGD3.33 million per month.


**Maximum Hypothetical Benefit**

*Cost of 1 month of deploying Wolbachia mosquitoes*: SGD 3.33mil 

*Hypothetical maximum benefit-cost ratio:*
```math
128.95 mil / 3.33 = 38.724 (8.7 times improvement from baseline!)
```






Overall, with our current prediction model and accounting for the error rate, the benefit-cost analysis of targeted deployment of project Wolbachia is much better compared to the baseline benefit-cost analysis.
