## 1. Introduction and Problem Statement

## Background

Dengue fever is a disease caused by the dengue virus and is a vector-borne disease whereby the virus is transmitted to human via the bite of an infected female _Aedes aegypti_ mosquito. The Aedes mosquito prefers to breed after rainy days where there are many pools of stagnant water. They also prefer the warm climate such as Singapore where it facilitates faster multiplication of the mosquito [(Reference)](https://www.nea.gov.sg/media/news/news/index/nea-urges-continued-vigilance-as-aedes-aegypti-mosquito-population-and-number-of-dengue-cases-remain-high-at-the-start-of-2020#:~:text=Climate%20change%20is%20anticipated%20to,the%20chances%20of%20dengue%20transmission.)

Once a person is bitten by an infected Aedes mosquito, there is an incubation period of between 4 to 7 days (ranges from 3 to 14 days)before a person displays symptoms of dengue fever which are sudden onset of fever, joint and muscle pain and skin rashes.  [(Reference)](https://www.moh.gov.sg/diseases-updates/dengue). 

Although dengue does not spread from person to person, the implications of dengue fever to the individual in terms of time off work and lost of productivity have an impact on healthcare burden and economic consequences. Time off work is also applicable to caregiver who may have to take time off to look after the infected individual. According to paper published in 2021 by Soh et. al, the average estimated economic impact of dengue in Singapore from 2010 to 2020 is approximately \\$1.014 to $2.265billion [(Reference)](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC10021432/#:~:text=We%20estimated%20that%20the%20average,21%2C262%20DALYs%20from%202010%E2%80%932020). 

In order to mitigate the economic cost of dengue fever, the National Environment Agency (NEA) has initiated several campaigns for dengue prevention programs. These initiatives include public awareness campaigns, community engagement efforts, and mosquito control measures. Despite these efforts, dengue outbreaks continue to pose a significant challenge to public health, burdening the healthcare system and affecting the quality of life for residents.

## Problem Statement 

The effectiveness of dengue prevention programs in Singapore is a critical concern for the National Environmental Agency. To this end, we will be **using machine learning models to predict the cases of dengue fever in Singapore using historical climate and google search trends.** Using metrics from the optimal modeling process, we will perform cost benefit analysis of dengue prevention program. This approach allows for evidence-based decision-making and optimization of resource allocation for public health initiatives.

### 2.1 Dengue Dataset

```python
#import
dengue = pd.read_csv('../data/dengue_data/WeeklyInfectiousDiseaseBulletinCases.csv')
```

```python
#check shape, and nulls and dtype
print(dengue.info())
dengue
```

It appears that there are no null values. However, the dataset includes all types of diseases - we want to do a row filter to get the dengue rows only.

#### 2.1.1 Filtering 'Dengue Fever'

```python
#check for all disease categories
dengue['disease'].unique()
```

```python
#slice dataframe for dengue fever only categories
dengue = dengue[dengue['disease'] == 'Dengue Fever']
dengue
```

#### 2.1.2 Fixing Dates

We need to fix the dates such that it matches the other DataFrames. Thus, we will reformat the 'epi_week' column into a typical YYYY-MM-DD format and convert it to datetime.

```python
#reformat time 
dates = []
for index, row in dengue.iterrows():
    epi_week = row['epi_week']
    year, week = epi_week.split('-W')
    date = datetime.datetime.strptime(f'{year}-W{week}-1', '%Y-W%U-%w').date()
    dates.append(date)
```

```python
#replace epi_week with reformatted time and reset index.
dengue['date'] = dates
dengue.drop(columns = 'epi_week', inplace = True)
dengue.reset_index(drop = True, inplace = True)
```

```python
#convert date format.
dengue['date'] = pd.to_datetime(dengue['date'])
print(dengue.info())
dengue
```

In order to simplify things, we will have the start dates of all the DataFrames be 1 January 2012.

```python
# Moving the date back by 1 day
dengue['date'] = dengue['date'] + timedelta(days=-1)
```

### 2.4 Weather Dataset

```python
#import
weather = pd.read_csv('../data/weather_data/master_weather_data.csv')
```

```python
print(weather.shape)
weather.info()
```

From the non-null counts, it seems like some casing error in the word 'min' on the source csv that led to the creation of two seperate columns. For "Highest 30 Min Rainfall (mm)" has 320004 rows and "Highest 30 min Rainfall (mm)" has 62707 which when added together gives 382711 rows which is the total rows. 

In anycase, the main objective to is the "Daily Rainfall Total (mm)" so the 6 columns will be dropped in favour of a cleaner dataframe. Additionally, the dataframe is not clean due to the expected numeric values being objects, so further cleaning is required.

#### 2.4.1 Dropping Unnecessary Columns

```python
weather.drop(columns = ['Highest 30 min Rainfall (mm)', 
                        'Highest 60 min Rainfall (mm)', 
                        'Highest 120 min Rainfall (mm)',
                        'Highest 30 Min Rainfall (mm)', 
                        'Highest 60 Min Rainfall (mm)', 
                        'Highest 120 Min Rainfall (mm)'
                       ], inplace = True)
```

```python
weather.info()
```

```python
#inspect weird elements.
print(weather['Mean Temperature (Â°C)'].unique()[:5])
print(weather['Daily Rainfall Total (mm)'].unique()[:5])
weather['Max Wind Speed (km/h)'].unique()
```

Seems like many missing datas has been imputed with a "â€”" or "-" causing the data to be handled as object. If the stations do not have the data, it can be safey removed as well. Therefore, converting to NaN and then dropping would be the most effective method without impact to this master data set.

#### 2.4.2 Dropping NaNs

```python
#replace and drop null values.
weather.replace('â€”', np.nan, inplace=True)
weather.replace('-', np.nan, inplace=True)
weather.dropna(inplace= True)
weather.reset_index(drop=True, inplace=True)
weather.info()
```

```python
#convert dtype from object to float for further processing.
for column in weather.columns:
    try:
        weather[column] = weather[column].astype(float)
    except:
        continue
weather.info()
```

```python
weather.columns.to_list()
```

#### 2.4.3 Fixing Dates

The dates are structured as 'Year', 'Month' and 'Day'. We will need to match it with the other DataFrames.

```python
#convert #separate year month dat into single date column
weather['Date'] = weather['Year'].astype(int).astype(str) + '/' + weather['Month'].astype(int).astype(str) + '/' + weather['Day'].astype(int).astype(str)

#convert new columns to date_time format
weather['Date'] = pd.to_datetime(weather['Date'])

#section off the data based on the other datsets available.
weather = weather[(weather['Year'] >= 2012) & (weather['Year'] <= 2022)]

#drop the individual dates into the proper columns.
weather.drop(columns = ['Year', 'Month', 'Day'], inplace = True)

#reorder columns
weather = weather[['Station',
                   'Date',
                   'Daily Rainfall Total (mm)',
                   'Mean Temperature (Â°C)',
                   'Maximum Temperature (Â°C)',
                   'Minimum Temperature (Â°C)',
                   'Mean Wind Speed (km/h)',
                   'Max Wind Speed (km/h)'
                  ]]
weather.reset_index(drop = True, inplace = True)
weather.info()
```

We will now use the mean number of all stations to get an average representation of the daily weather in Singapore.

```python
# Finding the mean of all stations
weather_sg = weather.groupby('Date').mean()
```

```python
weather_sg
```

Because our DataFrame has daily data while the other two only have weekly data, we will use the weekly mean of the weather data instead.

```python
weather_sg = weather_sg.resample('W').mean()
```

```python
weather_sg
```

```python
weather_sg = weather_sg.reset_index()
```

```python
weather_sg['Date'] = pd.to_datetime(weather_sg['Date'])
```

```python
weather_sg
```

```python
#export cleaned data
if not os.path.exists("../data/weather_data/output"):
    os.makedirs("../data/weather_data/output")
weather.to_csv("../data/weather_data/output/clean_weather_data.csv", index=False)
```

## 4. Exploratory Data Analysis

Before we do our modeling, it is important to understand our data. We will do that with our EDA.

```python
df.describe()
```

*Interpretation: Through the summary statistics, we can see a really high maximum value of dengue cases and Google search values as well. The daily rainfall has a high maximum value too.*

```python
sns.pairplot(data = df);
```

*Interpretation: We did a pairplot to have a quick understanding of the dataset. We see some strong trends between dengue cases and Google searches.*

```python
# Extracting year from the date for the box plots
df['year'] = pd.to_datetime(df['date']).dt.year

for i in df.columns[1:-1]:
    plt.figure(figsize=(10, 4))
    
    sns.boxplot(x=df['year'], y=df[i])
    

    # Modify the title to include the correlation coefficient
    plt.title(f'Distribution of {i.title()} over years')
    plt.xlabel('Year')
    plt.ylabel(i.title())
    
    plt.tight_layout()
    plt.show()
```

*Interpretation: This gives us an understanding of data distribution. We see a number of outliers for each column, but will not act on it for now.*

```python
for i in df.columns[2:-1]:
    plt.figure(figsize=(10, 4))
    
    sns.regplot(x=df[i], y=df['dengue_cases'], scatter_kws={'alpha':0.5})
    
    # Calculate the correlation coefficient
    correlation = df[i].corr(df['dengue_cases'])
    
    # Modify the title to include the correlation coefficient
    plt.title(f'Dengue Cases vs. {i.title()}\nCorrelation Coefficient: {correlation:.2f}')
    plt.xlabel(i.title())
    plt.ylabel('Dengue Cases')
    
    plt.tight_layout()
    plt.show()
```

*Interpretation: Now, we run a scatterplot between dengue cases and other columns. We see a strong positive relationship between dengue cases and Google searches, while there is little to no relationship between dengue cases and weather data.*

```python
fig = plt.figure(figsize=(12, 7))
fig.suptitle('Line Plots with Synchronized Axes: Relationship between Dengue Cases and Dengue Google Search')

# Plotting the first line
ax1 = sns.lineplot(x="date", y="dengue_cases", data=df)
ax1.grid(which='both', axis='x', visible=True, alpha=0.4)

# Setting y-axis label
ax1.set_ylabel('Dengue Cases')

# Setting secondary y-axis label for Google Search
ax2 = ax1.twinx()
ax2.set_ylabel('Google Search: Dengue')

# Plotting the second line (Google Search)
sns.lineplot(x="date", y="search_dengue", color='tab:orange', data=df, ax=ax2)

# Creating a third axes instance for the third line plot
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Adjust the position of the third y-axis
ax3.set_ylabel('Google Search: Dengue Symptoms')

# Plotting the third line (Google Search for Symptoms)
sns.lineplot(x="date", y="search_dengue_symptoms", color='tab:green', data=df, ax=ax3)

# Setting legends with labels
ax1.legend(['Dengue Cases'], loc="upper left")
ax2.legend(['Google Search: Dengue'], loc="upper right")
ax3.legend(['Google Search: Dengue Symptoms'], loc="upper center")

plt.show()
```

*Interpretation: To have an understanding of the relationship between dengue cases and the Google search trends, we use a lineplot with synchronised axes. We can see that there is a really strong relationship between the two.*

```python
plt.figure(figsize=(10, 4))
    
sns.regplot(data = df, x='search_dengue', y='search_dengue_symptoms', scatter_kws={'alpha':0.5})
    
# Calculate the correlation coefficient
correlation = df['search_dengue'].corr(df['search_dengue_symptoms'])
    
# Modify the title to include the correlation coefficient
plt.title(f'Google Search (Dengue) vs Google Search (Dengue Symptoms)\nCorrelation Coefficient: {correlation:.2f}')
plt.xlabel('Dengue')
plt.ylabel('Dengue Symptoms')
plt.show()
```

*Interpretation: Between the two search terms, we do see quite a strong relationship between 'dengue' and 'dengue symptoms' search terms.*

```python
fig, ax = plt.subplots(figsize=(2,6))

corr_matrix = df.corr(numeric_only = True)
corr_target = corr_matrix[['dengue_cases']].drop(labels=['dengue_cases'])

sns.heatmap(corr_target, vmin=-1, vmax=1, annot = True, cmap='coolwarm')
plt.title('Heatmap: Correlation scores between predictors and dengue cases', fontdict = {'size': 14}, loc = 'center');
```

*Interpretation: Finally, a heatmap shows us correlation scores between dengue cases and the other columns. Again, we see a strong relationship with Google search terms, with much lower scores for weather data.*
